# Assignments

The assignment of using the chemical composition data of wine to freely train your own classification model among several classifications to identify the type of wine and to check the accuracy of the model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##Data load and Preprocessing

```python
from sklearn.datasets import load_wine

wine = load_wine()
```

##Inspecting Data and Splitting Train_set, Test_set

* test_size = 0.3
* Check how the data is organized through print(wine['DESCR'])

```python
# Check the Dataset Descriptions
print(wine['DESCR'])
```

```python
print(wine.feature_names)
print(wine.target_names)
```

### Please define X and Y to predict `class` through `wine_df`.

```python
wine_df = pd.DataFrame(wine['data'], columns=wine.feature_names)
wine_df['class'] = wine.target

# Setting X, Y
X = wine_df.drop('class', axis=1)
Y = wine_df['class']
```

```python
print(X.head())
```

```python
print(Y.head())
print(Y.tail())
```

```python
from sklearn.preprocessing import StandardScaler
################### Freely Preprocess (Scaling MUST be done after splitting train/test!) ######################
```

### Split train and test data through `train_test_split`.
- test_size = 0.3, random_state = 0

```python
from sklearn.model_selection import train_test_split

# Splitting Train/Test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
# Scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

###### Freely Preprocessing 2 : Scaling ( After splitting; Data Leakage) #######
```

##Build Models Using Various Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# --- Logistic Regression Model ---
# Generate Model
lr_model = LogisticRegression(random_state=0, max_iter=200)
# Fitting
lr_model.fit(X_train_scaled, Y_train)
# Get predict value
Y_predict_lr = lr_model.predict(X_test_scaled)

# --- Decision Tree Classifier Model ---
# Generate Model
dt_model = DecisionTreeClassifier(random_state=0)
# Fitting
dt_model.fit(X_train_scaled, Y_train)
# Get predict value
Y_predict_dt = dt_model.predict(X_test_scaled)

# --- K-Nearest Neighbors (KNN) Classifier Model ---
# Generate Model
knn_model = KNeighborsClassifier(n_neighbors=5)
# Fitting
knn_model.fit(X_train_scaled, Y_train)
# Get predict value
Y_predict_knn = knn_model.predict(X_test_scaled)
```

### Logistic Regression Model Performance

```python
from sklearn.metrics import classification_report, accuracy_score

print("--- Logistic Regression Performance ---")
print(classification_report(Y_test, Y_predict_lr))
accuracy_lr = accuracy_score(Y_test, Y_predict_lr)
print("Accuracy: {0:.3f}".format(accuracy_lr))
```

### Decision Tree Classifier Performance

```python
from sklearn.metrics import classification_report, accuracy_score

print("--- Decision Tree Classifier Performance ---")
print(classification_report(Y_test, Y_predict_dt))
accuracy_dt = accuracy_score(Y_test, Y_predict_dt)
print("Accuracy: {0:.3f}".format(accuracy_dt))
```

### K-Nearest Neighbors (KNN) Classifier Performance

```python
from sklearn.metrics import classification_report, accuracy_score

print("--- K-Nearest Neighbors Classifier Performance ---")
print(classification_report(Y_test, Y_predict_knn))
accuracy_knn = accuracy_score(Y_test, Y_predict_knn)
print("Accuracy: {0:.3f}".format(accuracy_knn))
```
